```python
# wfdb - package for processing physiologic signal and annotation data
import wfdb
```

```python
# read a record
record = wfdb.rdrecord(record_name='data/ltdb/1.0.0/14046', sampfrom=1000 ,sampto=2000)
```

```python
# plotting ECG
wfdb.plot_wfdb(record=record, title='Record 14046 from ltdb')
display(record.__dict__)
```

```python
# reading signals as numpy array
signals, fields = wfdb.rdsamp('data/ltdb/1.0.0/14046', sampfrom=1000, sampto=2000)
display(signals)
display(fields)
```

```python
# Reading just annotation from record
annotation = wfdb.rdann('data/ltdb/1.0.0/14046', 'atr', sampfrom=1000, sampto=2000)
# annotation = wfdb.rdann('data/physionet.org/content/mitdb/1.0.0/x_mitdb/x_108', 'atr', sampfrom=1000, sampto=2000)
annotation.fs = 360
wfdb.plot_wfdb(annotation=annotation, time_units='minutes')
```

```python
wfdb.plot_wfdb(record=record, annotation=annotation,
               title='Record 100 from MIT-BIH Arrhythmia Database',
               time_units='seconds')
```
